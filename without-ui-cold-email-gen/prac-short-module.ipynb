```python
import chromadb
client = chromadb.Client()
collection = client.create_collection(name="my_collection")
```

```python
collection.add(
    documents = [
        "I want to drive a car",
        "The watch is old late 1950",
        "Watch your steps, there is a speed breaker ahead"
    ],
    ids = [
        "id1","id2","id3"
    ],
    metadatas = [
        {"url":"source1 url id1"},
        {"url":"source2 url id2"},        
        {"url":"source3 url id3"},
    ]
)
```

```python
collection.get(ids="id1")
```

```python
collection.query(query_texts="lets go on a shop, to buy clock", n_results=1)
```
